```python
from requests import get
import cv2
import gimpml
import matplotlib.pyplot as plt
import numpy as np
import torch
torch.cuda.is_available()
```

```python
url = 'https://th.bing.com/th/id/OIP.HRaK3uHyOSHgbcEc5N4yAAHaI5?w=853&h=1024&rs=1&pid=ImgDetMain'
if url != '':
  with open('/content/imagem_web.jpg','wb') as f:
    f.write(get(url).content)
```

```python
#arquivo = '/content/imagem.jpg'
arquivo = '/content/imagem_web.jpg'
```

```python
# Carregar a imagem original e a máscara
imagem = cv2.imread(arquivo)[:,:,::-1]
mascara = gimpml.parseface(imagem)

# Criar uma máscara binária onde a cor RGB (0,0,204) está presente
mascara_binaria = cv2.inRange(mascara, (0, 0, 204), (0, 0, 204))

# Definir a cor RGB
#cor_rgb = (255, 0, 0) # vermelho
cor_rgb = (30, 60, 210)
opacidade = 0.25

# Criar uma imagem colorida com a cor RGB vermelha
imagem_colorida = np.zeros_like(imagem)
imagem_colorida[mascara_binaria == 255] = cor_rgb

# Aplicar o blend apenas na região da máscara

imagem_final = imagem.copy()
imagem_final[mascara_binaria == 255] = cv2.addWeighted(imagem[mascara_binaria == 255], 1 - opacidade, imagem_colorida[mascara_binaria == 255], opacidade, 0)

# Exibindo
fig = plt.figure(figsize=(10, 7))
rows, columns=1,2
fig.add_subplot(rows, columns, 1)
plt.imshow(imagem)
plt.axis('off')
plt.title("Original")
fig.add_subplot(rows, columns, 2)
plt.imshow(imagem_final)
plt.axis('off')
plt.title("Alterada")
```

Teste com HSL

```python
# Carregar a imagem original e a máscara
imagem = cv2.imread(arquivo)[:,:,::-1]
mascara = gimpml.parseface(imagem)

# Criar uma máscara binária onde a cor RGB (0,0,204) está presente
mascara_binaria = cv2.inRange(mascara, (0, 0, 204), (0, 0, 204))

# Converter a imagem original para HSL
imagem_hsl = cv2.cvtColor(imagem, cv2.COLOR_BGR2HLS)


# Definir a cor HSL vermelha (200,0,0)
cor_hsl = np.uint8([[[0, 200, 200]]])  # H, L, S
cor_hsl = cv2.cvtColor(cor_hsl, cv2.COLOR_HLS2BGR)[0][0]

# Criar uma imagem colorida com a cor HSL vermelha
imagem_colorida = np.zeros_like(imagem_hsl)
imagem_colorida[mascara_binaria == 255] = cor_hsl

# Aplicar o blend apenas na região da máscara
opacidade = 0.15
imagem_resultante = imagem_hsl.copy()
imagem_resultante[mascara_binaria == 255] = cv2.addWeighted(imagem_hsl[mascara_binaria == 255], 1 - opacidade, imagem_colorida[mascara_binaria == 255], opacidade, 0)

# Converter a imagem de volta para BGR
imagem_final = cv2.cvtColor(imagem_resultante, cv2.COLOR_HLS2BGR)

# Exibindo
fig = plt.figure(figsize=(10, 7))
rows, columns=1,2
fig.add_subplot(rows, columns, 1)
plt.imshow(imagem)
plt.axis('off')
plt.title("Original")
fig.add_subplot(rows, columns, 2)
plt.imshow(imagem_final)
plt.axis('off')
plt.title("Alterada")
```
